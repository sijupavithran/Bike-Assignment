```python
import pandas as pd 
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
%matplotlib inline
```

```python
#Import ML Libraries

from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.preprocessing import MinMaxScaler

# Importing RFE and LinearRegression
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("/Users/I323638/Downloads/day.csv")
```

```python
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
climate = pd.get_dummies(df['season'])
```

```python
climate.head()
```

```python
climate.rename(columns={1:'spring',2:'summer',3:'fall',4:'winter'}, inplace=True)
```

```python
climate.drop(['spring'], axis = 1, inplace = True)
```

```python
df = pd.concat([df, climate], axis = 1)
```

```python
weather = pd.get_dummies(df['weathersit'])
```

```python
weather.head()
```

```python
weather.rename(columns={1:'clear',2:'mist',3:'light rain'}, inplace=True)
```

```python
weather.drop(['clear'], axis = 1, inplace = True)
```

```python
df = pd.concat([df, weather], axis = 1)
```

```python
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
seasons = pd.get_dummies(df['season'])
```

```python
seasons.head()
```

```python
seasons.rename(columns={1:'spring',2:'summer',3:'fall',4:'winter'}, inplace=True)
```

```python
seasons.drop(['spring'], axis = 1, inplace = True)
```

```python
bike_detail = pd.concat([df, seasons], axis = 1)
```

```python
bike_detail.head()
```

```python
#Dropping instant
bike_detail.drop(['instant'], axis = 1, inplace = True)
```

```python
#season and weathersit had to be dropeed as keeping these will cause R2 issues
bike_detail.drop(['season'], axis = 1, inplace = True)
bike_detail.drop(['weathersit'], axis = 1, inplace = True)
```

```python
weatherlist =  ['summer', 'fall', 'winter', 'mist', 'light rain']
```

```python
# Defining the map function
def binary_map(x):return x.map({True: 1, False: 0})
```

```python
bike_detail.head()
```

```python
# Check for duplicates
print(bike_detail.columns[bike_detail.columns.duplicated()].tolist())
```

```python
# Remove duplicate columns
bike_detail = bike_detail.loc[:, ~bike_detail.columns.duplicated()]
```

```python
bike_detail.head()
```

```python
# Applying the function to the housing list
bike_detail[weatherlist] = bike_detail[weatherlist].apply(binary_map)
```

```python
bike_detail.head()
```

```python
#Converting dteday to ddmmyy to extract month, date of week and to check if the extracted date is a weekend
bike_detail['dteday'] = pd.to_datetime(bike_detail['dteday'], format="%d-%m-%Y")
```

```python
# Create time features
bike_detail['month'] = bike_detail['dteday'].dt.month
bike_detail['dow'] = bike_detail['dteday'].dt.dayofweek
bike_detail['is_weekend'] = (bike_detail['dteday'].dt.dayofweek >= 5).astype(int)
```

```python
#Dropping date and month as required info are extracted from date
#dropping weekday as weekday and dow same
bike_detail.drop(['dteday'], axis = 1, inplace = True)
bike_detail.drop(['mnth'], axis = 1, inplace = True)
bike_detail.drop(['weekday'], axis = 1, inplace = True)
```

```python
bike_detail.head()
```

```python
#Casual and registered are dropped from the dataset as they are dependent variables to cnt. Keeping this will cause VIF issues
bike_detail.drop(['casual'], axis = 1, inplace = True)
bike_detail.drop(['registered'], axis = 1, inplace = True)
```

```python
bike_detail.head()
```

```python
#APPLY SCALING
```

```python
# the train and test data set should have same rows
np.random.seed(0)
df_train, df_test = train_test_split(bike_detail, train_size = 0.7, test_size = 0.3, random_state = 100)
```

```python
scaler = MinMaxScaler()
```

```python
# Apply scaler() to all the columns except the 'yes-no' and 'dummy' variables
num_vars = ['temp', 'atemp', 'hum', 'windspeed','cnt', 'month', 'dow']
df_train[num_vars] = scaler.fit_transform(df_train[num_vars])
```

```python
df_train.head()
```

```python
plt.figure(figsize = (16, 10))
sns.heatmap(df_train.corr(), annot = True, cmap="YlGnBu")
plt.show()
```

```python
sns.pairplot(df_train)
plt.show()
```

```python
y_train = df_train.pop('cnt')
X_train = df_train
```

```python
# Running RFE with the output number of the variable equal to 10
lm = LinearRegression()
lm.fit(X_train, y_train)
rfe = RFE(lm, n_features_to_select = 10)             # running RFE
rfe = rfe.fit(X_train, y_train)
```

```python
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
col = X_train.columns[rfe.support_]
col
```

```python
X_train.columns[~rfe.support_]
```

```python
X_train_rfe = X_train[col]
```

```python
# enabling constants
import statsmodels.api as sm  
X_train_rfe = sm.add_constant(X_train_rfe)
```

```python
lm = sm.OLS(y_train,X_train_rfe).fit()
```

```python
print(lm.summary())
```

```python
# Calculate the VIFs as all columns are significant
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = pd.DataFrame()
X = X_train_rfe
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

```python
y_train_price = lm.predict(X_train_rfe)
```

```python
# Plot the histogram of errors
fig = plt.figure()
sns.distplot((y_train - y_train_price), bins = 35)
fig.suptitle('Error Terms for the data', fontsize = 43)                  # Plot heading 
plt.xlabel('Error Info', fontsize = 24)         
plt.show() 
```

```python
X_train_new = X_train_rfe
X_train_new = X_train_new.drop(['const'], axis=1)
```

```python
#Apply Scaling
```

```python
# Apply scaler() to all the columns except the 'yes-no' and 'dummy' variables
num_vars = ['temp', 'atemp', 'hum', 'windspeed','cnt', 'month', 'dow']
df_test[num_vars] = scaler.fit_transform(df_test[num_vars])
```

```python
y_test = df_test.pop('cnt')
X_test = df_test
```

```python
# Creating X_test_new dataframe by dropping variables from X_test
X_test1 = X_test[X_train_new.columns]

# Adding a constant variable 
X_test1 = sm.add_constant(X_test1)
```

```python
# Making predictions
y_pred = lm.predict(X_test1)
```

```python
fig = plt.figure()
plt.scatter(y_test,y_pred)
fig.suptitle('y_test vs y_pred', fontsize=20)              # Plot heading 
plt.xlabel('y_test', fontsize=18)                          # X-label
plt.ylabel('y_pred', fontsize=16)                          # Y-label
plt.show() #to understand the spread
```

```python
#R2 Score
r2_score(y_test, y_pred)
```

```python
#Relative Impact of Features on Bike Demand
import matplotlib.pyplot as plt
import pandas as pd

# Create a dataframe from the regression coefficients
data = {
    "Feature": ["yr", "holiday", "temp", "hum", "windspeed", "summer", "fall", "winter", "mist", "light rain"],
    "Coefficient": [0.2307, -0.0865, 0.4961, -0.1390, -0.1827, 0.1168, 0.0746, 0.1630, -0.0537, -0.2409]
}
df = pd.DataFrame(data)

# Sort features by coefficient magnitude
df["AbsCoef"] = df["Coefficient"].abs()
df_sorted = df.sort_values("Coefficient", ascending=False)

# Plot
plt.figure(figsize=(10, 6))
plt.barh(df_sorted["Feature"], df_sorted["Coefficient"])
plt.xlabel("Coefficient Value (Impact on Demand)")
plt.ylabel("Feature")
plt.title("Relative Impact of Features on Bike Demand")
plt.axvline(0, color="black", linewidth=0.8)
plt.gca().invert_yaxis()
plt.show()
```

```python
#Effect of Categorical Variables on Bike Demand
# Create a summary DataFrame for categorical effects
cat_effects = pd.DataFrame({
    "Category": [
        "Year (2019 vs 2018)", 
        "Holiday (Yes vs No)", 
        "Season: Summer vs Spring", 
        "Season: Fall vs Spring", 
        "Season: Winter vs Spring", 
        "Weather: Mist vs Clear", 
        "Weather: Light Rain vs Clear"
    ],
    "Effect on Demand": [0.2307, -0.0865, 0.1168, 0.0746, 0.1630, -0.0537, -0.2409]
})

# Sort for better visualization
cat_effects = cat_effects.sort_values("Effect on Demand", ascending=True)

# Plot
plt.figure(figsize=(10, 6))
plt.barh(cat_effects["Category"], cat_effects["Effect on Demand"])
plt.xlabel("Impact on Demand (Coefficient Value)")
plt.title("Effect of Categorical Variables on Bike Demand")
plt.axvline(0, color="blue", linewidth=0.8)
plt.tight_layout()
plt.show()
```
